```python
import pandas as pd
```

Read the csv file to a dataframe that contains all NSS scores for teaching institutions of full-time students:

```python
chunks = pd.read_csv("teach_ft.csv", chunksize=100_000)
NSS = pd.concat(chunks, ignore_index=True)
```

Examine the data:

```python
NSS.head()
```

```python
NSS.info()
```

Reduce down to only the columns necessary for the site (see Data Dictionary.csv for full details): 

```python
NSS = NSS[['UKPRN', 'PROVIDER_NAME', 'LEVEL_OF_STUDY', 'SUBJECT_LEVEL', 'CAH_CODE',
       'CAH_NAME', 'QUESTION_NUMBER', 'NUMBER_RESPONSES', 'NUMBER_POPULATION',
       'SUPPRESSION_REASON', 'POSITIVITY_MEASURE', 'BENCHMARK', 'DIFFERENCE',
       'PUB_RESPONSE_HEADCOUNT', 'PUB_RESPRATE']]
```

```python
NSS.info()
```

Clean - replace NULLs in text fields with "N/A"s, etc:

```python
NSS['CAH_CODE'].fillna("N/A", inplace=True)
NSS['CAH_NAME'].fillna("N/A", inplace=True)
NSS.info()
```

```python
NSS.info()
```

Keep only the rows for the "Themes", not individual questions, and keep only "All undergraduates" under LEVEL_OF_STUDY:

```python
NSS = NSS[NSS["QUESTION_NUMBER"].str.contains("Theme")]
NSS = NSS[NSS["LEVEL_OF_STUDY"].str.contains("All undergraduates")]
```

```python
NSS.info()
```

Export to csv for further analysis in Excel:

```python
NSS.to_csv("NSS_themes.csv", index=False)
```
